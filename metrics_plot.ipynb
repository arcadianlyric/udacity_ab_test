```python
# set color palette
col = [(0.12156862745098039, 0.4666666666666667, 0.7058823529411765),
 (1.0, 0.4980392156862745, 0.054901960784313725),
 (0.17254901960784313, 0.6274509803921569, 0.17254901960784313),
 (0.8392156862745098, 0.15294117647058825, 0.1568627450980392),
 (0.5803921568627451, 0.403921568627451, 0.7411764705882353),
 (0.5490196078431373, 0.33725490196078434, 0.29411764705882354),
 (0.8901960784313725, 0.4666666666666667, 0.7607843137254902),
 (0.4980392156862745, 0.4980392156862745, 0.4980392156862745),
 (0.7372549019607844, 0.7411764705882353, 0.13333333333333333),
 (0.09019607843137255, 0.7450980392156863, 0.8117647058823529)]

def plot_metrics(df):
    fig, ax = plt.subplots(figsize=(15, 6))
    ax.plot(df['Date'], df['Pageviews'], color=col[0], label='Pageviews')
    ax.plot(df['Date'], df['Clicks'], color=col[1], label='Clicks')
    ax.plot(df['Date'], df['Enrollments'], color=col[2], label='Enrollments')
    ax.plot(df['Date'], df['Payments'], color=col[3], label='Payments')
    ax.legend()
    ax.set_xlabel('Date', fontsize=12)
    ax.set_ylabel('Metrics', fontsize=12)
    plt.show()
```

```python
plot_metrics(ctl)
```

```python
plot_metrics(exp)
```
